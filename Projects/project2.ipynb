#1.Finite difference:
This part include 3 type difference but they all used same idea df/dx=(f(x)-f(x-deltax)) and their difference is that we can choose different start point and their tend direction, Taking the abscissa x_0 as the origin, the forward difference is a process of moving in the positive direction of the x-axis, the backward difference is a process of moving in the negative direction of the x-axis, and the central difference is an process based on x_0 as the midpoint, and then select the left and right sides as starting point and slowly approaching x_0.

we use an sample function f(x)=x^2



```python
import matplotlib.pyplot as plt
import numpy as np

#set up function
f = lambda x: x**2 

#set up how many points are displayed in the image here I select start at 0 and end at 5, and between them we have 100 number
x_min, x_max = 0, 5
npoints = 100
x0 = (x_min+x_max)/2
deltax = 0.5

#draw plot
x = np.linspace(x_min, x_max, npoints)
plt.plot(x,f(x))

#draw the derivative
x_1 = np.array([x0-deltax, x0, x0+deltax])
plt.plot(x_1,f(x_1),'-o')
for point in x_1:
    linex, liney = [point, point], [0, f(point)]
    plt.plot(linex, liney, color='black', linewidth=2.0)    

plt.text(x_1[0]+deltax/2, f(x_1[0]+deltax/2)+1, r'backward difference', {'ha': 'left', 'va': 'bottom'}, rotation=90)    
plt.text(x_1[1]+deltax/2, f(x_1[1]+deltax/2)+1, r'forwward difference', {'ha': 'left', 'va': 'bottom'}, rotation=90)    

    
plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([0,5])
plt.show()
```

#2 Complex step
The complex step derivative approximation is a technique to compute the derivative of a real-valued function f(x).

For example：origin function f(x)=x^2
1. we subsititude x+ih(here h is constant and i is imaginary unit) to x, we get f(x+ih)=(x+ih)(x+ih)=x^2-h^2+2ihx we found real part is x^2-h^2 and imaginary part would be 2x*ih
2. we find out f'(x)=2x
3. compare f'(x) and 2x*ih we can find their difference is 2x

now we use easy linear tayler serious f(x+ih)=f(x)+ihf'(x)

f(x)=f(x)=x+3

```python
#complex step 
k lambda x:x+3
def f_imag(x):
    #in order to make it simple calculate I mark h as 1
    x_0=x+1j
    f1=(k(x_0)-k(x))/(1j)
    return(f1)

#finite difference
def f_infi(x,y):
    #y is delta x
    f2=(k(x+y)-k(x))/y
    return(f2)

x_1=np.linspace(1,100,100)


#draw plot x and difference f'(x)
plt.plot(x_1,f_imag(x_1),'-r')
plt.plot(x_1,f_infi(x_1,0.5),'-b')
plt.plot(x_1,f_infi(x_1,5),'-g')
plt.plot(x_1,f_infi(x_1,100),'-y')
plt.legend([ 'complex step','finite difference','finite difference','finite difference'])
plt.show()

#When the deltax gradually becomes smaller, the images of the two methods gradually overlap. 
#According to the definition of the derivative, when the deltax is small enough, the derivative is more accurate, 
#so we can show that the complex-order function is effective
```

we use same sample,for integral f(x)/(x-z)^(n+1)*dx,we have(here our main problem is how to find integral, so we can used integral calcultor to confirm it)

```python
import matplotlib.pyplot as plt
import numpy as np

def cauchy(n,z):
    #n is n time derivative, z is what we want to find derivative x value,x use 5
    integral_part=-2*z*np.math.log(abs(10+z))-z**2/(10+z)+5
    imag_part=np.math.factorial(n)/((2*np.math.pi)*1j)
    f3=integral_part*imag_part
    return f3
y_list=[]
x_1=np.linspace(1,100,100)
for z in x_1:
    y_list.append(cauchy(1,z)/1j)

plt.plot(x_1,y_list,'-r')
plt.plot(x_1,f_infi(x_1,0.5),'-b')
plt.plot(x_1,f_imag(x_1),'-y')
```
